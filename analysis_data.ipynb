```python
# reformate the features 
def features_converter(obj):
    res=ast.literal_eval(obj)
    return res
```

```python
df['features']=df['features'].apply(features_converter)
```

```python
df['features'] = df['features'].apply(lambda x: [''] if len(x) < 2 else x).apply(lambda x:''.join(x))
```

```python
#reformate the description 
def descripiton_conveter(obj):
    a=ast.literal_eval(obj)
    return a
```

```python
df['description']=df['description'].apply(descripiton_conveter)
```

```python
df.description=df.description.apply(lambda x:' ' if len(x)<=2 else x).apply(lambda x:''.join(x))
```

```python
#reformate the categories 

def categories_converter(obj):
    a=ast.literal_eval(obj)
    return a
```

```python
df.categories=df.categories.apply(categories_converter)
```

```python
df.categories=df.categories.apply(lambda x:' ' if len(x)<2 else x).apply(lambda x:''.join(x))
```
