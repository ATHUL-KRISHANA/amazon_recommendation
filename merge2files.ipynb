```python
import pandas as pd

# Define the path to your CSV files
large_csv_file_path = 'output_file1.csv'
small_csv_file_path = 'amazon_headphone.csv'

# Specify the chunk size
chunk_size = 1000000  # Adjust the chunk size based on your memory capacity

# Load the smaller CSV file (left DataFrame)
small_data = pd.read_csv(small_csv_file_path)
print("Small CSV file loaded successfully.")

# Create an empty DataFrame to store merged results
merged_data_chunks = pd.DataFrame()

# Process the large CSV file in chunks and perform left join
for chunk in pd.read_csv(large_csv_file_path, chunksize=chunk_size):
    # Perform the left join with the small_data DataFrame
    merged_chunk = pd.merge(small_data, chunk, how='inner', on='parent_asin')

    # Append the merged chunk to the merged_data_chunks DataFrame
    merged_data_chunks = pd.concat([merged_data_chunks, merged_chunk])

# Now you have the entire large CSV file merged with the small CSV file
print("Data merged successfully.")

# Display the first few rows of the merged data
print(merged_data_chunks.head())
```
