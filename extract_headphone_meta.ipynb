```python
headphone=[]
for i in df.title:
    if 'headset' in str(i).lower() or 'headphone' in str(i).lower() or 'headphones' in str(i).lower():
        headphone.append('yes')
    else:
        headphone.append('no')
```

```python
len(headphone)
```

```python
df['if_headset']=headphone
```

```python
df
```

```python
#extract the headphone dataframe
headphone_meta=df.loc[df['if_headset']=='yes']
```

```python
headphone_meta.reset_index(inplace=True)
```

```python
headphone_meta
```

```python
headphone_meta.to_csv('amazon_headphone.csv')
```
